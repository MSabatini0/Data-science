```python
import csv

seq_ADN =[] #lista de listas con cada secuencia de ADN
with open('secuencias ADN.txt') as file:
    reader = csv.reader(file)
    for row in reader:
        seq_ADN.append(row)

print(seq_ADN[0][0][1])
```

```python
gc = [] # valores de  %GC para cada secuencia en forma de lista

for row in seq_ADN:
    count = 0
    for string in row:
        for letter in string:
            if letter in 'GgCc':
                count +=1
        valor = count/len(string)
        gc.append(valor*100)

print(gc)
```

```python
import csv

seq_prot =[] #lista de listas con cada secuencia de ADN
with open('secuencias proteinas b.txt') as file:
    reader = csv.reader(file)
    for row in reader:
        seq_prot.append(row)

print(len(seq_prot))
```

```python
alif = [] # valores de  %Alif para cada secuencia en forma de lista

for row in seq_prot:
    count = 0
    for string in row:
        for letter in string:
            if letter in 'LlIiVv':
                count +=1
        valor = count/len(string)
        alif.append(valor*100)

print(alif)
print(len(alif))
```

```python
polares = [] # valores de  %polares para cada secuencia en forma de lista

for row in seq_prot:
    count = 0
    for string in row:
        for letter in string:
            if letter in 'DdEeKkRr':
                count +=1
        valor = count/len(string)
        polares.append(valor*100)

print(polares)
```

```python
cys = [] # valores de  %cys para cada secuencia en forma de lista

for row in seq_prot:
    count = 0
    for string in row:
        for letter in string:
            if letter in 'Cc':
                count +=1
        valor = count/len(string)
        cys.append(valor*100)

print(cys)
```

```python
prol = [] # valores de  %prol para cada secuencia en forma de lista

for row in seq_prot:
    count = 0
    for string in row:
        for letter in string:
            if letter in 'Cc':
                count +=1
        valor = count/len(string)
        prol.append(valor*100)

print(prol)
```

```python
arom = [] # valores de  %arom para cada secuencia en forma de lista

for row in seq_prot:
    count = 0
    for string in row:
        for letter in string:
            if letter in 'FfYyWw':
                count +=1
        valor = count/len(string)
        arom.append(valor*100)

print(arom)
```

```python
import pandas as pd

datos = pd.read_excel('datos proteinas b.xlsx')
#datos['%GC gen'] = gc
datos['%alif'] = alif
datos['%cys'] = cys
datos['%arom'] = arom
datos['%prol'] = prol

sol = datos['Solubilidad_pH=8(0/1)']
#exp = datos['expresion(1-5)']
datos= datos.drop (columns= ['Solubilidad_pH=8(0/1)'])

datos['Solubilidad_pH=8(0/1)'] = sol

datos.to_excel('datos proteinas biblio - train.xlsx', index=False)
```
